# Data Visualization

the ausdex module comes with several data visulisations out of the box build with [plotly](https://plotly.com/python/).

Visualizations can be created via the command line and saved as html files, or created programmatically.

Inflation data can be visulised by a time series of the values of a given price inflated to a particular "dollar year". Additionally, the consumer price index (CPI) time series can be plotted.

## Consumer Price Index (CPI) over time
The time series of the CPI can be created as an html file via the command line or programmatically via the API.

Using the following commands, a time series fo the CPI can be generated via the following command:

```bash
ausdex plot-cpi [OPTIONS]
```

With the following options:
  
- `--show` / `--no-show`         Whether or not to show the figure in a browser.  [default: show]
- `--output`                    The path to where the figure will be saved. Output can be PDF, JPG, PNG or HTML based on the extension.
- `--start-date`                Date to set the beginning of the time series
                                graph. Defaults to None, which starts in
                                1948.
- `--end-date`                  Date to set the end of the time series graph
                                too. If empty, then the end date to the most
                                recent quarter.
- `--location` The location for calculating the CPI. 
    Must be one of: Australia, Sydney, Melbourne, Brisbane, Adelaide, Perth, Hobart, Darwin, Canberra. Default: Australia. 
- `--title`                     A custom title of the plot.
- `--help`                          Show the help message.
  
The same plot can be made programmatically via the API. More information found in the [API docs](reference.md):

````python
from ausdex.inflation import plot_cpi_timeseries
fig_cpi = plot_cpi_timeseries(start_date=1960, end_date='12-12-2016')
fig_cpi.show()
````

## Equivalent dollar value over time
An inflation chart can be created and saved as an html file via the following command:

```bash
ausdex plot-inflation [OPTIONS] COMPARE_DATE
```

With the following arguments and options:

- `COMPARE_DATE`: Date to set relative value of the dollars too.     

Options:

-  `--show` / `--no-show`         Whether or not to show the figure in a
                                  browser.  [default: show]
- `--output`: The path to where the figure will be saved.
                                  Output can be PDF, JPG, PNG or HTML based on
                                  the extension.
- `--start-date`: Date to set the beginning of the time series
                                  graph. Defaults to None, which starts in
                                  1948.
- `--end-date`: Date to set the end of the time series graph
                                  too. If empty, then the end date to the most
                                  recent quarter.    
- `--location` The location for calculating the CPI. 
    Must be one of: Australia, Sydney, Melbourne, Brisbane, Adelaide, Perth, Hobart, Darwin, Canberra. Default: Australia. 
- `--value`: Value you in `compare_date` dollars to plot on the time series. Defaults to 1.
- `--help`: Shows the help

The same chart can be created with the same arguments programmatically. More information found in the [API docs](reference.md):

````python
from ausdex.inflation import plot_inflation_timeseries
fig = plot_inflation_timeseries('12-12-2019', start_date='12-12-1950', end_date='12-06-2020')
fig.show()
````
